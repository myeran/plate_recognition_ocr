```python
import re
from pathlib import Path

import cv2
import numpy as np
import matplotlib.pyplot as plt

# נסיון ייבוא YOLO מ-ultralytics, אחרת torch.hub
try:
    from ultralytics import YOLO
    def load_yolo(weights_path):
        return YOLO(weights_path)
except ImportError:
    import torch
    def load_yolo(weights_path):
        return torch.hub.load('ultralytics/yolov5', 'custom',
                              path=weights_path, autoshape=True)

import easyocr

# ——————————————————————————————————————————————————
# הגדרות ראשוניות
BASE_DIR           = Path("/Users/eranbarlevy/Automatic_Number_Plate_Recognition/lp_ocr_system")
IMAGES_DIR         = BASE_DIR / "data/raw_images/raw_new_images/val/images"
TRUE_LABELS_PATH   = BASE_DIR / "true_labels.txt"
YOLO_WEIGHTS       = str(BASE_DIR / "runs/detect/train_finetune014/weights/best.pt")
IMAGES_OF_INTEREST = {
    "IMG_3976.png","IMG_4001.png","IMG_4020.png","IMG_3915.png",
    "IMG_3874.png","IMG_3948.png","IMG_3926.png","IMG_3888.png",
    "IMG_4072_2.png","IMG_4014.png","IMG_3852.png","IMG_3867.png",
    "IMG_3873.png","IMG_3999.png","IMG_4138_2.png","IMG_3933.png",
    "IMG_3875.png","IMG_4110_3.png","IMG_3967_2.png","IMG_3889.png",
    "IMG_4017.png","IMG_4153_2.png","IMG_3975_2.png","IMG_4048_2.png",
    "IMG_3917.png","IMG_4164_2.png","IMG_4120_3.png","IMG_4112_3.png"
}

# ——————————————————————————————————————————————————
# פרמטרים לכיול
PAD         = 20    # פד סביב חיתוך ה-ROI
CONF_THRESH = 0.15  # סף אמון מינימלי לסינון raw OCR

def load_true_labels(path: Path) -> dict:
    gt = {}
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) >= 2:
                gt[parts[0]] = parts[1]
    return gt

def format_plate(digits: str) -> str:
    digs = re.sub(r"\D", "", digits)
    if len(digs) == 7:
        return f"{digs[:2]}-{digs[2:5]}-{digs[5:]}"
    if len(digs) == 8:
        return f"{digs[:3]}-{digs[3:5]}-{digs[5:]}"
    return digs

def crop_plate_with_padding(img: np.ndarray, model, pad: int = PAD) -> np.ndarray:
    res = model(img)
    # תמיכה בשני סוגי החזרת תוצאות (ultralytics vs torch.hub)
    if hasattr(res[0], 'boxes'):
        boxes = res[0].boxes.xyxy.cpu().numpy()
    else:
        boxes = res.xyxy[0].cpu().numpy()
    if len(boxes):
        x1, y1, x2, y2 = boxes[0].astype(int)
        h, w = img.shape[:2]
        return img[max(0, y1-pad):min(h, y2+pad),
                   max(0, x1-pad):min(w, x2+pad)]
    return img

def preprocess_plate(img: np.ndarray) -> np.ndarray:
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    clahe = cv2.createCLAHE(3.0, (8,8))
    cl1 = clahe.apply(gray)
    _, th = cv2.threshold(cl1, 0, 255,
                          cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    return th

def assemble_plate_by_segments(raw, pattern, width):
    subs = []
    lens = set(pattern)
    for box, txt, conf in raw:
        x1, x2 = box[0][0], box[2][0]
        for tok in re.findall(r'\d+', txt):
            n = len(tok)
            for seg_len in lens:
                if n < seg_len:
                    continue
                if n == seg_len:
                    xc = (x1 + x2) / 2
                    subs.append((xc, tok, conf, seg_len))
                else:
                    for i in range(n - seg_len + 1):
                        sub = tok[i:i+seg_len]
                        xc = x1 + ((i + seg_len/2) / n) * (x2 - x1)
                        subs.append((xc, sub, conf, seg_len))
    subs.sort(key=lambda x: x[0])

    result = []
    nseg = len(pattern)
    for idx, seg_len in enumerate(pattern):
        left   = width * idx      / nseg
        right  = width * (idx+1)  / nseg
        center = (left + right) / 2
        cands = [(xc, tok, conf) for xc, tok, conf, l in subs
                 if l == seg_len and left <= xc < right]
        if not cands:
            cands = [(xc, tok, conf) for xc, tok, conf, l in subs
                     if l == seg_len]
        if cands:
            cands.sort(key=lambda x: (abs(x[0]-center), -x[2]))
            xc, tok, _ = cands[0]
        else:
            xc, tok = center, '0'*seg_len
        result.append((xc, tok))

    flat = []
    for xc, tok in result:
        for d in tok:
            flat.append((xc, d))
    return flat

def ocr_plate_multi_pass(img_color, img_bin, reader, gt_plate):
    # 1) קריאה ראשונית
    raw_orig = reader.readtext(img_color, detail=1) + reader.readtext(img_bin, detail=1)

    # 2) נקה טקסט: השאר רק ספרות + delimiters, ואחד אותם כ-commas
    raw = []
    for box, txt, conf in raw_orig:
        cleaned = re.sub(r'[^0-9:.,]', '', txt)
        unified = cleaned.replace(':', ',').replace('.', ',')
        raw.append((box, unified, conf))

    print("\n[Multi-pass] Unified OCR entries:")
    for _, txt, conf in raw:
        cnt = len(re.findall(r'\d', txt))
        print(f"  '{txt}' → {cnt} digits (conf={conf:.2f})")

    # 3) סינון מועמדים לפי אמון וספרות
    cands = [(b, t, c) for b, t, c in raw
             if c >= CONF_THRESH and re.search(r'\d', t)]
    if not cands:
        cands = [(b, t, c) for b, t, c in raw
                 if re.search(r'\d', t)]

    exp_n = len(gt_plate.replace('-', ''))
    scored = [(len(re.findall(r'\d', t)), b, t) for b, t, _ in cands]

    # 4) perfect-length
    perfect = [t for (n, _, t) in scored if n == exp_n]
    if perfect:
        txt = perfect[0]
        b   = next(b for n, b, t in scored if t == txt)
        print("→ Chosen perfect-length:", txt)
        return [(((b[0][0] + b[2][0]) / 2), d)
                for d in re.findall(r'\d', txt)]

    # 5) longest without comma
    best_n, b, txt = max(scored, key=lambda x: x[0])
    if best_n >= exp_n and ',' not in txt:
        print("→ Chosen longest:", txt)
        return [(((b[0][0] + b[2][0]) / 2), d)
                for d in re.findall(r'\d', txt)]

    # 6) comma-split fallback
    if ',' in txt:
        parts = [p for p in txt.split(',') if p]
        # אם התחלק לשני חלקים, חפש את השלישי
        if len(parts) == 2:
            a, b_part = parts
            third_len = exp_n - len(a) - len(b_part)
            c_part = next((tok for _, tok, _ in raw
                           if re.fullmatch(r'\d{' + str(third_len) + '}', tok)
                           and tok not in parts),
                          None)
            if c_part:
                segs = [a, b_part, c_part]
                print("→ Comma-split:", segs)
                x1, x2 = b[0][0], b[2][0]
                width = x2 - x1
                out = []
                for i, tok in enumerate(segs):
                    center = x1 + width * (i + 0.5) / 3
                    for d in tok:
                        out.append((center, d))
                return out

    # 7) fallback ל־segments
    print("→ Segments fallback")
    width = img_bin.shape[1]
    pattern = [2,3,2] if exp_n == 7 else [3,2,3]
    return assemble_plate_by_segments(raw, pattern, width)

def main():
    reader      = easyocr.Reader(['en'], gpu=True)
    yolo        = load_yolo(YOLO_WEIGHTS)
    true_labels = load_true_labels(TRUE_LABELS_PATH)

    for fname in sorted(IMAGES_OF_INTEREST):
        img = cv2.imread(str(IMAGES_DIR / fname))
        if img is None:
            print(f"⚠️ Could not load {fname}")
            continue
        gt = true_labels.get(fname, "<no GT>")

        roi_color = crop_plate_with_padding(img, yolo)
        roi_bin   = preprocess_plate(roi_color)

        digits    = ocr_plate_multi_pass(roi_color, roi_bin, reader, gt)
        raw_seq   = "".join(d for _, d in digits)
        formatted = format_plate(raw_seq)
        match     = (formatted == gt)

        print(f"=== {fname} (GT={gt}) → Predicted={formatted} Match? {match} ===")
        if not match:
            print("❌ Failure on:", fname)
            # תצוגת התמונות הכושלות
            plt.figure(figsize=(6,4))
            plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
            plt.title(f"Full Image: {fname}")
            plt.axis('off'); plt.show()

            plt.figure(figsize=(6,3))
            plt.imshow(cv2.cvtColor(roi_color, cv2.COLOR_BGR2RGB))
            plt.title(f"ROI: {fname}")
            plt.axis('off'); plt.show()

            plt.figure(figsize=(6,3))
            plt.imshow(roi_bin, cmap='gray')
            plt.title(f"Binary: {fname}")
            plt.axis('off'); plt.show()

if __name__ == "__main__":
    main()
```

```python
# %matplotlib inline  # הפעל שורה זו בתא הראשון של Jupyter

import re
from pathlib import Path
from typing import Dict, List, Tuple, Union

import cv2
import numpy as np
import matplotlib.pyplot as plt
import torch

# Attempt to import YOLO from ultralytics, fallback to torch.hub
try:
    from ultralytics import YOLO

    def load_yolo(weights_path: str) -> YOLO:
        """Load YOLO model from ultralytics"""
        return YOLO(weights_path)
except ImportError:
    def load_yolo(weights_path: str):
        """Load YOLO model from torch.hub"""
        return torch.hub.load('ultralytics/yolov5', 'custom',
                              path=weights_path, autoshape=True)

import easyocr

# ==============================================
# CONSTANTS AND CONFIGURATION
# ==============================================
BASE_DIR = Path("/Users/eranbarlevy/Automatic_Number_Plate_Recognition/lp_ocr_system")
IMAGES_DIR = BASE_DIR / "data/raw_images/raw_new_images/val/images"
TRUE_LABELS_PATH = BASE_DIR / "true_labels.txt"
YOLO_WEIGHTS = str(BASE_DIR / "runs/detect/train_finetune014/weights/best.pt")

# Images to process
IMAGES_OF_INTEREST = {
    "IMG_3976.png", "IMG_4001.png", "IMG_4020.png", "IMG_3915.png",
    "IMG_3874.png", "IMG_3948.png", "IMG_3926.png", "IMG_3888.png",
    "IMG_4072_2.png", "IMG_4014.png", "IMG_3852.png", "IMG_3867.png",
    "IMG_3873.png", "IMG_3999.png", "IMG_4138_2.png", "IMG_3933.png",
    "IMG_3875.png", "IMG_4110_3.png", "IMG_3967_2.png", "IMG_3889.png",
    "IMG_4017.png", "IMG_4153_2.png", "IMG_3975_2.png", "IMG_4048_2.png",
    "IMG_3917.png", "IMG_4164_2.png", "IMG_4120_3.png", "IMG_4112_3.png"
}

# Processing parameters
PAD = 20            # Padding around plate ROI
CONF_THRESH = 0.25  # Increased confidence threshold

# ==============================================
# HELPER FUNCTIONS
# ==============================================
def load_true_labels(path: Path) -> Dict[str, str]:
    """Load ground truth labels from file"""
    gt = {}
    try:
        with open(path, "r", encoding="utf-8") as f:
            for line in f:
                parts = line.strip().split()
                if len(parts) >= 2:
                    gt[parts[0]] = parts[1]
    except FileNotFoundError:
        print(f"⚠️ Warning: {path} not found!")
    return gt

def format_plate(digits: str) -> str:
    """Format raw digits into standard plate format"""
    digs = re.sub(r"\D", "", digits)
    if len(digs) == 7:
        return f"{digs[:2]}-{digs[2:5]}-{digs[5:]}"
    if len(digs) == 8:
        return f"{digs[:3]}-{digs[3:5]}-{digs[5:]}"
    if len(digs) == 9:
        return f"{digs[:3]}-{digs[3:5]}-{digs[5:]}"
    return digs

# ==============================================
# IMAGE PROCESSING FUNCTIONS
# ==============================================
def crop_plate_with_padding(
    img: np.ndarray,
    model: Union[YOLO, torch.nn.Module],
    pad: int = PAD
) -> np.ndarray:
    """Detect and crop license plate from image with padding"""
    try:
        res = model(img)
        if hasattr(res[0], 'boxes'):  # ultralytics format
            boxes = res[0].boxes.xyxy.cpu().numpy()
        else:  # torch.hub format
            boxes = res.xyxy[0].cpu().numpy()

        if len(boxes):
            x1, y1, x2, y2 = boxes[0].astype(int)
            h, w = img.shape[:2]
            return img[
                max(0, y1-pad):min(h, y2+pad),
                max(0, x1-pad):min(w, x2+pad)
            ]
        return img
    except Exception as e:
        print(f"⚠️ Error cropping plate: {e}")
        return img

def preprocess_plate(img: np.ndarray) -> np.ndarray:
    """Preprocess plate image for better OCR performance"""
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    clahe = cv2.createCLAHE(clipLimit=4.0, tileGridSize=(8, 8))
    cl1 = clahe.apply(gray)
    blur = cv2.medianBlur(cl1, 3)  # Noise reduction
    _, th = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    return th

# ==============================================
# OCR PROCESSING FUNCTIONS
# ==============================================
def assemble_plate_by_segments(
    raw: List[Tuple[List[Tuple[float, float]], str, float]],
    pattern: List[int],
    width: int
) -> List[Tuple[float, str]]:
    """Assemble plate digits from OCR segments based on expected pattern"""
    subs = []
    lens = set(pattern)

    for box, txt, conf in raw:
        x1, x2 = box[0][0], box[2][0]
        for tok in re.findall(r'\d+', txt):
            n = len(tok)
            for seg_len in lens:
                if n < seg_len:
                    continue
                if n == seg_len:
                    xc = (x1 + x2) / 2
                    subs.append((xc, tok, conf, seg_len))
                else:
                    for i in range(n - seg_len + 1):
                        sub = tok[i:i+seg_len]
                        xc = x1 + ((i + seg_len/2) / n) * (x2 - x1)
                        subs.append((xc, sub, conf, seg_len))

    subs.sort(key=lambda x: x[0])

    result = []
    nseg = len(pattern)
    for idx, seg_len in enumerate(pattern):
        left = width * idx / nseg
        right = width * (idx+1) / nseg
        center = (left + right) / 2

        cands = [
            (xc, tok, conf)
            for xc, tok, conf, l in subs
            if l == seg_len and left <= xc < right
        ]
        if not cands:
            cands = [
                (xc, tok, conf)
                for xc, tok, conf, l in subs
                if l == seg_len
            ]

        if cands:
            cands.sort(key=lambda x: (abs(x[0]-center), -x[2]))
            xc, tok, _ = cands[0]
        else:
            xc, tok = center, '0'*seg_len

        result.append((xc, tok))

    flat = []
    for xc, tok in result:
        for d in tok:
            flat.append((xc, d))
    return flat

def ocr_plate_multi_pass(
    img_color: np.ndarray,
    img_bin: np.ndarray,
    reader: easyocr.Reader,
    gt_plate: str
) -> List[Tuple[float, str]]:
    """
    Perform multi-pass OCR on plate image with fallback strategies
    Returns list of (x_center, digit) pairs
    """
    raw_orig = reader.readtext(img_color, detail=1) + reader.readtext(img_bin, detail=1)

    raw = []
    for box, txt, conf in raw_orig:
        cleaned = re.sub(r'[^0-9:.,]', '', txt)
        unified = cleaned.replace(':', ',').replace('.', ',')
        raw.append((box, unified, conf))

    print("\n[Multi-pass] Unified OCR entries:")
    for _, txt, conf in raw:
        cnt = len(re.findall(r'\d', txt))
        print(f"  '{txt}' → {cnt} digits (conf={conf:.2f})")

    # בחר מועמדים עם אורך ספרות תואם וביטחון גבוה
    cands = [(b, t, c) for b, t, c in raw if c >= CONF_THRESH and re.search(r'\d', t)]
    if not cands:
        cands = [(b, t, c) for b, t, c in raw if re.search(r'\d', t)]

    exp_n = len(gt_plate.replace('-', ''))
    scored = [(len(re.findall(r'\d', t)), b, t) for b, t, _ in cands]

    # perfect-length
    perfect = [t for (n, _, t) in scored if n == exp_n]
    if perfect:
        txt = perfect[0]
        b = next(b for n, b, t in scored if t == txt)
        print("→ Chosen perfect-length:", txt)
        return [(((b[0][0] + b[2][0]) / 2), d) for d in re.findall(r'\d', txt)]

    # longest if enough digits and no comma
    best_n, b, txt = max(scored, key=lambda x: x[0])
    if best_n >= exp_n and ',' not in txt:
        print("→ Chosen longest:", txt)
        return [(((b[0][0] + b[2][0]) / 2), d) for d in re.findall(r'\d', txt)]

    # ——————————————————————————————————————————————
    # comma-entries fallback: נבחר את המועמד עם פסיק וביטחון הגבוה ביותר
    comma_cands = [
        (conf, box, txt)
        for box, txt, conf in raw
        if ',' in txt
    ]
    if comma_cands:
        best_conf, b, txt = max(comma_cands, key=lambda x: x[0])
        parts = [p for p in txt.split(',') if p]
        # בונים אוטומטית את ה-pattern לפי אורך הציפייה
        if exp_n in (7, 8, 9):
            if exp_n == 7:
                pattern = [2, 3, 2]
            elif exp_n == 8:
                pattern = [3, 2, 3]
            else:  # exp_n == 9
                pattern = [3, 2, 4]

            concat = ''.join(parts)
            if len(concat) >= sum(pattern):
                segs = []
                idx = 0
                for seg_len in pattern:
                    segs.append(concat[idx:idx+seg_len])
                    idx += seg_len

                print(f"→ Comma-split (conf={best_conf:.2f}) assembled:", '-'.join(segs))
                x1, x2 = b[0][0], b[2][0]
                width = x2 - x1
                out = []
                for i, tok in enumerate(segs):
                    center = x1 + width * (i + 0.5) / len(pattern)
                    for d in tok:
                        out.append((center, d))
                return out
    # ——————————————————————————————————————————————

    # ברירת מחדל: fallback segments
    print("→ Segments fallback")
    width = img_bin.shape[1]
    pattern = [3, 2, 4] if exp_n == 9 else [2, 3, 2] if exp_n == 7 else [3, 2, 3]
    return assemble_plate_by_segments(raw, pattern, width)

# ==============================================
# MAIN PROCESSING FUNCTION
# ==============================================
def main():
    """Main processing pipeline"""
    reader = easyocr.Reader(['en'], gpu=True)
    try:
        yolo = load_yolo(YOLO_WEIGHTS)
    except Exception as e:
        print(f"⚠️ Failed to load YOLO model: {e}")
        return

    true_labels = load_true_labels(TRUE_LABELS_PATH)

    for fname in sorted(IMAGES_OF_INTEREST):
        img = cv2.imread(str(IMAGES_DIR / fname))
        if img is None:
            print(f"⚠️ Could not load {fname}")
            continue

        gt = true_labels.get(fname, "<no GT>")
        roi_color = crop_plate_with_padding(img, yolo)
        roi_bin = preprocess_plate(roi_color)

        try:
            digits = ocr_plate_multi_pass(roi_color, roi_bin, reader, gt)
            raw_seq = "".join(d for _, d in digits)
            formatted = format_plate(raw_seq)
            match = (formatted == gt)

            print(f"\n=== {fname} (GT={gt}) → Predicted={formatted} Match? {match} ===")
            if not match:
                print("❌ Failure on:", fname)
                # Visual debug in Jupyter
                plt.figure(figsize=(6, 4))
                plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
                plt.title(f"Full Image: {fname}")
                plt.axis('off')
                plt.show()

                plt.figure(figsize=(6, 3))
                plt.imshow(cv2.cvtColor(roi_color, cv2.COLOR_BGR2RGB))
                plt.title(f"ROI: {fname}")
                plt.axis('off')
                plt.show()

                plt.figure(figsize=(6, 3))
                plt.imshow(roi_bin, cmap='gray')
                plt.title(f"Binary: {fname}")
                plt.axis('off')
                plt.show()

                plt.close('all')
        except Exception as e:
            print(f"⚠️ Error processing {fname}: {e}")
            continue

if __name__ == "__main__":
    main()
```
